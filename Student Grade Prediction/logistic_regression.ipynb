```python
import pandas as pd
df = pd.read_csv('student-mat.csv')
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import sklearn.preprocessing as skpp
from sklearn.model_selection import train_test_split
import sklearn.model_selection as le_ms
import sklearn.metrics as sklm
import numpy as np
```

```python
data = df[["G1", "G2", "G3", "studytime", "failures", "absences", "sex", "internet"]]

#data_with_dummies = pd.get_dummies(data)
```

```python
data['G3_pass_fail'] = ((data['G3'] >= 10).replace({True: 1, False: 0}))
data['sex_binary'] = ((data['sex'] == 'M').replace({True: 1, False: 0}))
data['internet_binary'] = ((data['internet'] == 'yes').replace({True: 1, False: 0}))

data
```

```python
#data_with_dummies = data_with_dummies.drop(columns = ['G3']) # removing the G3 variable and reassigning the new data frame to the data with dummies name
data = data.drop(columns = ['G3', 'sex', 'internet']) # removing the G3 variable and reassigning the new data frame to the data with dummies name
data
```

```python
X = data.drop(columns= ['G3_pass_fail']).copy() # make a copy of only the x variables
y = data['G3_pass_fail'] # isolate the outcome (y variable)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1234, train_size=0.8, stratify= y)

columns = X_train.columns
```
